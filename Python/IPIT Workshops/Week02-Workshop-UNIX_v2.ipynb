# Introduction to UNIX

Most operating systems can trace their heritage back to the invention of UNIX in the Bell Labratories in the 1960's. The concept of UNIX was to design small programs that do one thing, and do that one thing well. Each of these smaller programs could then be used in combination to do much larger operations.

During your introduction to programming, you won't be required to do these larger complex operations. However, given modern operating systems such as Linux, MacOS, Android, iOS, and  Chrome OS are all built on UNIX like heritage it is useful to understand these roots.

## Special Directories

We used a special shortcut in one of the commands above. In UNIX based systems the following short key notation can be used to refer to where you are in the filesystem.

* dot or <code>.</code>refers to the current directory. This is the directory you are in at the moment.
* dot dot or <code>..</code> referes to the directory which is the parent of the current directory. You can think of this as the back button in a GUI filesystem. 
* tilde or <code>~</code> is a shortcut which refers to your home directory. For example, for user01, the file /home/user01/example.txt can also be denoted by ~/example.txt.

However, earlier we said that everything in the UNIX system is either a folder or a file. These shortcuts are no different. The <code>~</code>, <code>.</code> and <code>..</code> are simply folders in the filesystem that point to specific places just like any other folder does.

# Demo 5: File Ownership - chmod 

Remember the string of letters we saw at the start of each entry when running the <code>ls -l</code> command to populate our file listing?

It looked something like this:
<code>drwxr-xr-x</code>

What do these characters mean?

These characters indicate the *permissions* set for the particular file or folder. We already know that the first character shows if the entry is a directory <code>d</code> or a file <code>-</code>. The next three characters denote permissions <code>rwx</code> in three groups: User, Group and Other User rights.

<code>rwx</code> denotes what permissions each group has.

* r - read, given a numerical value of 4.
* w - write, given a numerical value 2.
* x - executable, with a numerical value of 1.

To set permissions we use the command <code>chmod</code> with a numerical value assigned to each group with respect to the permissions we wish to set. The first number referes to the current user, ie. you. The second group is the users group, since a user can belong to a larger group of users (think students vs staff on our university mainframe) and the final group is the others group, that is everyone else who is not either the current user or the users group. 

To assign permissions we add the numerical value of each permission we want to set together to get the final value.

For example, let's say I had a bash script (a small program) called <code>finalGrades.sh</code> I had written to calculate your final grade. I would want to be able to read and write to this file to be able to save new versions and execute it to run the program. This means I would need all three permissions; read (4), write (2) and execute (1) ( 4+2+1 =7). Thus, for the first group I would use the value 7. 

I might also want my academic staff colleagues to be able to read the script to ensure everything is being calculated as expected and run the script, but, not edit it: read (4) and execute (1) (4+1 = 5). 

Finally, for the professional staff (everyone else) who can access the staff mainframe I might just want them to be able to run the script at the end of the semester. This would be permission execute (1).

To set these permissions I would use the chmod command:

<code> chmod 751 finalGrades.sh </code>

<code>-rwxr-x--x   1 richard  staff        7368 18 Dec 14:02 finalGrades.sh</code>

Let's take a look.

Running <code>ls -l</code> you should find the file <code>finalGrades.sh</code>
